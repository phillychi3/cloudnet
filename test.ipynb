```python
import cloudnet.cuda
cloudnet.cuda.device()

from cloudnet.layers import *
from cloudnet.cloudnet import cloudet
import time
import numpy as np
import Fer2013 as fer


```

```python
(x_train,t_train),(x_test,t_test) = fer.load_fer(True, False, True, False, np.float32)
print(x_train, t_train)
print(x_train.shape, t_train.shape)
```

```python
model =  [
    Conv2D(32,3,1,0),
    Conv2D(32,3,1,0),
    MaxPool2D(2,2,2),
    Conv2D(64,3,1,0),
    Conv2D(64,3,1,0),
    MaxPool2D(2,2,2),
    Conv2D(128,3,1,0),
    Conv2D(128,3,1,0),
    MaxPool2D(2,2,2),

    flatten(),
    Dense(10),
]



net = cloudet(model)
net.setup(optimizer="Adam",activation="relu",shape=(1,48,48))
```

```python
batch_size = 30
train_size = x_train.shape[0]
iter_per_epoch = max((train_size//batch_size), 1)


for j in range(10):
  start = time.time()
  print("│====================================│  ")
  
  loss_t = 0
  for i in range(1,iter_per_epoch+1):
    batch_mask = np.random.choice(train_size, batch_size)                 #Random choose data

    x_batch = x_train[batch_mask]
    t_batch = t_train[batch_mask]
    
    loss = net.train(x_batch, t_batch)              #Train&Caculate the loss of the net
    loss_t += loss
    if i%10==0:
      print('│Iters {:<6} Loss        : {:<8} │  '.format(i,str(loss)[:8]), end='\r', flush=True)
  
  cost = time.time()-start
  test_acc = net.accuracy(x_test, t_test, batch_size)
  
#Save the parameters
  
  print("│Epoch {:<5} Average Loss : {:<8} │  ".format(j+1, str(loss_t/iter_per_epoch)[:8]))
  print("│            Test Acc     : {:<8} │  ".format(str(test_acc*100)[:8]))
  print("│            Cost Time    : {:<8} │  ".format(str(cost)[:8]))
  print("│            Iters/sec    : {:<8} │  ".format(str(iter_per_epoch/cost)[:8]))
    
print('└────────────────────────────────────┘  ')
```
